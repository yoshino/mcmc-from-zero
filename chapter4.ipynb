# Chapter4 メトロポリス法

```python
import random
import numpy as np
import matplotlib.pyplot as plt
```

### 4.2 期待値の計算の具体例

```python
# s(action): 作用
# 符号が逆であることを除いて統計学の対数尤度と同じ
def calc_4_2(s, niter):
  step_size = 0.5

  x = 0
  xs = []
  naccept = 0

  for i in range(niter):
    backup_x = x
    action_init = s(x)

    dx = random.uniform(-step_size, step_size)
    x += dx
    action_fin = s(x)
    
    # メトロポリステスト
    metropolis = random.uniform(0, 1)
    if (np.exp(action_init - action_fin) > metropolis):
      # 受理
      naccept += 1
      xs.append(x)
    else:
      # 棄却
      x = backup_x

  accept_ratio = naccept / niter
  
  return accept_ratio, xs
```

```python
s = lambda x: x**2 / 2
niter = 100000
accept_ratio, xs = calc_4_2(s, niter)
print('accept ratio: ', accept_ratio)
plt.hist(xs, bins=20)
```

### 4.4 ガウス分布以外の例

```python
# step_sizeをわたせるように修正しただけ
def calc_4_4(s, niter, step_size):
  x = 0
  xs = []
  naccept = 0

  for i in range(niter):
    backup_x = x
    action_init = s(x)

    dx = random.uniform(-step_size, step_size)
    x += dx
    action_fin = s(x)
    
    # メトロポリステスト
    metropolis = random.uniform(0, 1)
    if (np.exp(action_init - action_fin) > metropolis):
      # 受理
      naccept += 1
      xs.append(x)
    else:
      # 棄却
      x = backup_x

  accept_ratio = naccept / niter
  
  return accept_ratio, xs
```

#### ２つのガウス分布を重ね合わせた例

```python
p = lambda x: (np.exp(-(x-3)**2 / 2) + np.exp(-(x+3)**2)) / (2*np.sqrt(2*np.pi))
s = lambda x: - np.log(p(x))

# ステップサイズが0.5だと１つの山になってしまう
niter = 1000
accept_ratio, xs = calc_4_4(s, niter, 0.5)
print('accept ratio: ', accept_ratio)
plt.hist(xs, bins=100)
```

```python
# ステップサイズを大きくすると２つの山がサンプルできる
niter = 1000
accept_ratio, xs = calc_4_4(s, niter, 5)
print('accept ratio: ', accept_ratio)
plt.hist(xs, bins=100)
```

#### 半円とガウス分布

```python
# ２つのガウス分布を重ね合わせた例
def p(x):
  if x >=0:
    return np.exp(-x**2/2) / np.sqrt(2*np.pi)
  elif -1 <= x and x < 0:
    return np.sqrt(1-x**2) * (2/np.pi) 
  else:
    return 0

s = lambda x: - np.log(p(x))

niter = 100000
accept_ratio, xs = calc_4_2(s, niter)
print('accept ratio: ', accept_ratio)
plt.hist(xs, bins=100)
```
